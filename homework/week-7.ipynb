```python
# Load pandas and the data
import pandas as pd
import numpy as np
ride = pd.read_csv("ride_sharing.csv")
```

```python
# First few rows in ride_sharing.csv
ride.head()
```

```python
# show how many rows and columns are in this dataset
ride.shape # rows = 25760; columns = 10
```

```python
# All data types within the dataframe
ride.dtypes
```

```python
# The first column, "duration" is seen as a string.
# Using two snippets of code, clean this column up so that
# is now an integer (int64)
ride['duration'] = ride['duration'].str.strip('minutes')
ride['duration'] = ride['duration'].astype('int')
```

```python
# Now display the datatypes in the updated dataframe
ride.dtypes
```

```python
# Show the min and max values of ride durations (the shortest and longest rides)
# Hint: If 'df' is the name of your dataframe, simply specify the column name
# and then append the .min() or .max() methods.
print("Shortest ride duration:", ride['duration'].min())
print("Longest ride duration:", ride['duration'].max())
```

```python
# Create a new dataframe object and populate it with rides that are 60 minutes or less.
# Hint: Use the .query() method for the dataframe that we used in class.
short_rides = ride.query('duration <= 60')
```

```python
# Display created short_rides data frame
short_rides
```

```python
# Now show both the min and max values of the duration column for your new dataset.
print("Shortest ride duration (from the short_rides dataframe):", short_rides['duration'].min())
print("Longest ride duration (from the short_rides dataframe):", short_rides['duration'].max())
```

```python
# Using the new dataframe, update it so that it ONLY has records where
# the tire_size is 26.
short_rides = short_rides.query('tire_size == 26')
short_rides
```

```python
# New data frame with only FEMALE riders using the 26-inch wheeled rental bikes for their rentals of one hour or less
females_short_rides = short_rides.query('user_gender == "Female"')
females_short_rides
```

```python
# Finally, produce a count of all records with FEMALE riders
# using the 26-inch wheeled rental bikes for their rentals of
# one hour or less.
len(females_short_rides)
```
